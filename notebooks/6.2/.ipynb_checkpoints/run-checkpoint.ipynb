```python
from multiprocessing.pool import Pool
import requests
import json
from matplotlib import pyplot as plt
MAX_CON = 50
plt.style.use('seaborn-whitegrid')
```

```python
def run_function(url = 'https://j4fohrjqqyhud7giienbqoozxa0jrbyj.lambda-url.us-east-2.on.aws/'):
        ret = requests.get(url)
        return json.loads(ret.text)
```

```python
def run_exp(con=50):
    p = Pool(con)
    result = p.map(run_function, ['https://j4fohrjqqyhud7giienbqoozxa0jrbyj.lambda-url.us-east-2.on.aws/'] * con)
    return result
```

```python
raw_data = []

for i in range(1, MAX_CON + 1):
    print('current round is: {}'.format(i))
    tmp = run_exp(i)
    tmp = run_exp(i)
    raw_data.append(tmp)
```

```python
with open('4096MB_rawdata.txt', 'w') as f:
    json.dump(raw_data, f)
```

```python
def calculate_metric(m, func):
    total = 0
    for i in range(20):
        total += func(m[i])
    return total/20
```

```python
def parse_raw_data(raw_data, func):
    result = []
    for i in range(1, MAX_CON + 1):
        current_trail = raw_data[i-1]
        current_copy = 0
        for j in range(i):
            # there are 20 round of data
            metrics = current_trail[j]['stream_metric']
            current_copy += calculate_metric(metrics, func)
        current_copy = current_copy / i
        result.append(current_copy)
    return result

def parse_raw_data_runtime(raw_data):
    result = []
    for i in range(1, MAX_CON + 1):
        current_trail = raw_data[i-1]
        current_copy = 0
        for j in range(i):
            # there are 20 round of data
            metrics = current_trail[j]['runtime']
            current_copy += metrics
        current_copy = current_copy / i
        result.append(current_copy)
    return result
```

```python
best_copy_avg = parse_raw_data(raw_data, lambda i: i[0][0])
```

```python
x = list(range(1, MAX_CON+1))
```

```python
plt.title('Best Copy speed (8192MB)')
y = parse_raw_data(raw_data, lambda i: i[0][0])
plt.xlabel('Concurrency')
plt.ylabel('MB/S')
plt.plot(x, y)
```
