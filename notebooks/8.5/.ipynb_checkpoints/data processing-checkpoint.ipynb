```python
import json
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
import datetime
import os
import scipy.stats as stats



def get_cpustealdelta_per_min(data):
    return float(data['cpuStealDelta']) / (float(data['runtime']) / 60000)

def get_pagefaultdelta_per_min(data):
    return float(data['pageFaultsDelta']) / (float(data['runtime']) / 60000)

plt.style.use('seaborn-whitegrid')
```

```python


def get_jsons_from_dir(con):
#     base_dir = '../SAAF/test/history/cache_bench_test-cb_exp_{}-0MBs-run0-1/'
#     base_dir = '../6.22/data/cache_bench_test-cb_exp_{}-0MBs-run0-1/' # 4.5 GB
    base_dir = '../7.1/data/cache_bench_test-cb_exp_{}-0MBs-run0-1/' # 2.2 GB
#     base_dir = '../7.8/data/cache_bench_test-cb_exp_1-0MBs-run0-{}/'
#     base_dir = '../7.22/data/cache_bench_test-cb_exp_1-0MBs-run0-{}/' # computation

#     base_dir = '../8.5/computation_data/computation_test-cb_exp_{}-0MBs-run0-1/' # large array random access
    cur_dir = base_dir.format(con)
    file_names = os.listdir(cur_dir)
    results = []
    for fname in file_names:
        with open(cur_dir + fname, 'r') as f:
            results.append(json.load(f))
    return results
```

```python
raw_data = list(map(get_jsons_from_dir, range(1, 31)))
# raw_data = list(map(get_jsons_from_dir, range(1, 28))) # computation data (page fault)
# raw_data = list(map(get_jsons_from_dir, range(0, 30)))
```

```python
all_uuid = set()
all_container = set()
uuid_dict = dict()
container_dict = dict()
counter_uuid = 1
counter_container = 1
for d in raw_data:
    for c in d:
        if c['containerID'] not in all_container:
            all_container.add(c['containerID'])
        if c['uuid'] not in all_uuid:
            all_uuid.add(c['uuid'])
        if c['uuid'] not in uuid_dict:
            uuid_dict[c['uuid']] = counter_uuid
            counter_uuid += 1
        if c['containerID'] not in container_dict:
            container_dict[c['containerID']] = counter_container
            counter_container += 1
container_count = len(uuid_dict)
```

```python
def get_outlier_info(con):
    y_target = y_arr[x_arr == con]
#     y_target
    outlier_dots = np.array(raw_data[con-1])[y_target > 2 * y_target.mean()]
#     print(np.where(np.any(y_target > 400, axis=0)))
    print('uuid:')
    for i in outlier_dots:
        print(uuid_dict[i['uuid']])
#         print(i['uuid'])
    print('containerID:')
    for i in outlier_dots:
        print(container_dict[i['containerID']])
```
